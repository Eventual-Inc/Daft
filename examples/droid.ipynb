```python

import daft
from daft import col
```

```python
droid_meta = daft.datasets.droid.raw()

droid_meta.show(3, format="fancy", max_width=40)
```

```python
# Show some statistics from the metadata 
droid_meta_stats = droid_meta.agg_list(
    col("success"),
)
```

```python
droid_traj = daft.datasets.droid.trajectory(droid_meta)

droid_traj.show(3, format="fancy", max_width=40)
```

```python
droid_video = daft.datasets.droid.camera_frames(droid_traj)

droid_video.show(3, format="fancy", max_width=40)
```
