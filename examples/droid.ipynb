```python
import daft
from daft import col
```

```python
droid_raw = daft.datasets.droid.raw().limit(10)

droid_raw.show(10, format="fancy", max_width=40)
```

```python
droid_traj = daft.datasets.droid.trajectory(droid_raw)

droid_traj.show(3, format="fancy", max_width=40)
```

```python
droid_traj.agg(col("wrist_cam_video").is_null())
```

```python
hdf5_attrs = droid_raw.select(col("trajectory").hdf5_attrs())

hdf5_attrs.show(3, format="fancy", max_width=40)
```

```python
hdf5_keys = droid_raw.select(col("trajectory").hdf5_keys())

hdf5_keys.show(3, format="fancy", max_width=40)
```

```python
hdf5_meta = droid_raw.select(col("trajectory").hdf5_metadata())

hdf5_meta.show(3, format="fancy", max_width=40)
```

```python
droid_traj = daft.datasets.droid.trajectory(droid_raw)

droid_traj.show(3, format="fancy", max_width=40)
```

```python
droid_video = daft.datasets.droid.camera_frames(droid_traj)

droid_video.show(3, format="fancy", max_width=40)
```
